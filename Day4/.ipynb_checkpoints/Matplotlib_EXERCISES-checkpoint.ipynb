**1. Prepare scatter plot with given data (x,y). Try to use both approaches (stateful, stateles). Your graph should include title, xlabel, ylabel and grid. Save your graph as my_plot.pdf.**

```python
import matplotlib.pyplot as plt
```

```python
# stateful approach

x = [1,2,3,4,5,7,8,9,10,11]
y = [1,3,5,8,1,6,3,6,8,9]
plt.plot(x)
plt.plot(y)
plt.title('I am title')
plt.xlabel('I am x label')
plt.ylabel('I am y label')
plt.grid(True)
plt.show()
#plt.savefig('my_plot_stateful.pdf')
```

```python
# stateless approach

x = [1,2,3,4,5,7,8,9,10,11]
y = [1,3,5,8,1,6,3,6,8,9]

fig, ax = plt.subplots()
ax.plot(x)
ax.plot(y)
ax.grid(True)
ax.set(title='I am title', xlabel='I am x label', ylabel='I am y label')
plt.show()
#plt.savefig('my_plot_stateless.pdf')
```

**2. Try to make same graph as in ex2.pdf with given data (x1,y1,x2,y2)**

```python
x1 = [1,2,3,4,5,7,8,9,10,11]
y1 = [1,3,5,8,1,6,3,6,8,9]
x2a = [1, 0, 2, 4, 6]
x2b = [1, 2, 3, 3, 4]

# plt.figure(1)
plt.subplot(121) # two rows, 1 column and first plot
plt.plot(x1, y1, 'bo')
plt.grid(True)
plt.title('Some title...')
plt.xlabel('xlabel')
plt.ylabel('ylabel')

plt.subplot(122) # two rows, 1 column and second plot
plt.plot(x2a, 'r--', label='I am boring', color='blue')
plt.plot(x2b, 'k', label='legend', color='red')
plt.title('I am title!')
plt.xlabel('This is the X axis')
plt.ylabel('This is the Y axis')
plt.xticks([1, 2, 3, 4, 5])
plt.legend()
plt.show()
#plt.savefig('my_ex2.pdf')
```
